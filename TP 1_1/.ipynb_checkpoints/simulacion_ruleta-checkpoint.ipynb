```python
def graficar_varianza(dict_tiradas, esPromedio):
    fig, ax = plt.subplots()

    ax.set_xlabel('n (número de tiradas)')
    ax.set_ylabel('vv (valor de la varianza)')

    for dataframe in dict_tiradas.values():
        ax.plot(dataframe["df"]["varianza"], label="vvn (valor de la varianza del número X con respecto a n)")
    
    ax.axhline(varianza_esperada, color="black", label="vve (valor de la varianza esperada)")
    
    #las dos siguientes lineas son para mostrar el tick de la freciencia relativa esperada en el grafico
    trans = transforms.blended_transform_factory(ax.get_yticklabels()[0].get_transform(), ax.transData)
    ax.text(0, varianza_esperada, "{:.1f}".format(varianza_esperada), color="black", transform=trans, ha="right", va="center")

    plt.legend()
        
    if esPromedio:
        ax.set_title(f"VARIANZA PROMEDIO {cantidad_corridas} CORRIDAS")
        plt.savefig(f"img/varianza-promedio-{cantidad_corridas}-corridas.png")
    else:
        ax.set_title(f"VARIANZA {cantidad_corridas} CORRIDAS")
        plt.savefig(f"img/varianza-{cantidad_corridas}-corridas.png")
        
    plt.show()
```

```python
def graficar_desvio_estandar(dict_tiradas, esPromedio):
    fig, ax = plt.subplots()    
    
    ax.set_xlabel('n (número de tiradas)')
    ax.set_ylabel('vd (valor del desvío)')

    for dataframe in dict_tiradas.values():
        ax.plot(dataframe["df"]["desvio_estandar"], label="vd (valor del desvío del número X)")
    
    ax.axhline(desviacion_estandar_esperada, color="black", label="vde (valor del desvío esperado)")
    
    #las dos siguientes lineas son para mostrar el tick de la freciencia relativa esperada en el grafico
    trans = transforms.blended_transform_factory(ax.get_yticklabels()[0].get_transform(), ax.transData)
    ax.text(0, desviacion_estandar_esperada, "{:.1f}".format(desviacion_estandar_esperada), color="black", transform=trans, ha="right", va="center")

    plt.legend()
        
    if esPromedio:
        ax.set_title(f"DESVÍO ESTANDAR PROMEDIO {cantidad_corridas} CORRIDAS")
        plt.savefig(f"img/desviacion-estandar-esperada-promedio-{cantidad_corridas}-corridas.png")
    else:
        ax.set_title(f"DESVÍO ESTANDAR {cantidad_corridas} CORRIDAS")
        plt.savefig(f"img/desviacion-estandar-esperada-{cantidad_corridas}-corridas.png")
    
    plt.show()
```

```python
def graficar_histograma_frecuencia(frecuencias_por_corrida, esPromedio):
    fig, ax = plt.subplots()
    
    ax.set_xlabel("n")
    ax.set_ylabel("fr (frecuencia relativa)")
    
    for dataframe in frecuencias_por_corrida:
        ax.bar(dataframe.index, dataframe["fr%"], alpha=0.5)

    ax.axhline((1/37)*100, color="black", label="fre (frecuencia relativa esperada)")

    #las dos siguientes lineas son para mostrar el tick de la freciencia relativa esperada en el grafico
    trans = transforms.blended_transform_factory(ax.get_yticklabels()[0].get_transform(), ax.transData)
    ax.text(0,(1/37)*100, "{:.1f}".format((1/37)*100), color="black", transform=trans, ha="right", va="center")

    ax.legend()
    
    if esPromedio:
        ax.set_title(f"HISTOGRAMA PROMEDIO {cantidad_corridas} CORRIDAS")
        plt.savefig(f"img/histograma-frecuencia-relativa-promedio{cantidad_corridas}-corridas.png")
    else:
        ax.set_title(f"HISTOGRAMA {cantidad_corridas} CORRIDAS")
        plt.savefig(f"img/histograma-frecuencia-relativa-{cantidad_corridas}-corridas.png")
        
    plt.show()
```

```python
def generar_dataframe_tiradas():
    tiradas = np.random.randint(0,37,size=[cantidad_tiradas])

    df_tiradas = pd.DataFrame(data=tiradas, columns=["valor"])
    return df_tiradas
```

```python
def generar_dataframe_frecuencias():
    frecuencia_absoluta = pd.value_counts(df_tiradas["valor"])

    #df_frecuencias = pd.DataFrame({"valor":frecuencia_absoluta.index, "fa": frecuencia_absoluta.values})
    df_frecuencias = pd.DataFrame()
    df_frecuencias["valor"] = frecuencia_absoluta.index
    df_frecuencias["fa"] = frecuencia_absoluta.values
    df_frecuencias["fr"] = df_frecuencias["fa"]/len(df_tiradas)
    df_frecuencias["fr%"] = 100*df_frecuencias["fr"]
    df_frecuencias.set_index("valor", inplace = True)
    # ordenamos la lista de frecuencias segun los valores para facilitar su manejo
    return df_frecuencias
```

```python
def estadisticas_segun_tiradas():
    fr_segun_cantidad_tiradas = []
    fa_segun_cantidad_tiradas = []
    cantidad_ocurrencias = 0
    numero = np.random.randint(0,37)  # generamos un numero aleatorio para su estudio
    esperanza_segun_cantidad_tiradas = []
    varianza_segun_cantidad_tiradas = []
    desvio_segun_cantidad_tiradas = []
    tiradas = []

    for index, row in df_tiradas.iterrows():
        if row["valor"] == numero:  # si el valor de la tirada actual (tirada.valor) es igual al numero de estudio, aumentamos en 1 la cantidad de ocurrencias de ese numero
            cantidad_ocurrencias += 1
            fa_segun_cantidad_tiradas.append(cantidad_ocurrencias)

        fr_segun_cantidad_tiradas.append(100*cantidad_ocurrencias/(index+1))   # calculamos la frecuencia relativa actual y la guardamos en una lista

        tiradas.append(row["valor"])  # guardamos el valor de la tirada actual en una nueva lista
        esperanza_segun_cantidad_tiradas.append(np.mean(tiradas))   # calculamos la esperanza actual
        varianza_segun_cantidad_tiradas.append(np.var(tiradas))   # calculamos la varianza actual
        desvio_segun_cantidad_tiradas.append(np.std(tiradas))   # calculamos la desviacion estandar actual
    
    return fa_segun_cantidad_tiradas, fr_segun_cantidad_tiradas, esperanza_segun_cantidad_tiradas, varianza_segun_cantidad_tiradas, desvio_segun_cantidad_tiradas, numero
    
```

# INICIO

```python
cantidad_tiradas = 5000
cantidad_corridas = 10

dict_tiradas = {}
frecuencias_por_corrida = []

for corrida in range(cantidad_corridas):
    print(f"CORRIDA NÚMERO {corrida+1}")
    
    df_tiradas = generar_dataframe_tiradas()

    frecuencias_por_corrida.append(generar_dataframe_frecuencias())
    
    fa_segun_cantidad_tiradas, fr_segun_cantidad_tiradas, esperanza_segun_cantidad_tiradas, varianza_segun_cantidad_tiradas, desvio_segun_cantidad_tiradas, numero_estudio = estadisticas_segun_tiradas()
    df_tiradas["frecuencia_relativa"] = fr_segun_cantidad_tiradas
    df_tiradas["esperanza"] = esperanza_segun_cantidad_tiradas
    df_tiradas["varianza"] = varianza_segun_cantidad_tiradas
    df_tiradas["desvio_estandar"] = desvio_segun_cantidad_tiradas
    
    datos_tirada = {
        "numero_estudio": numero_estudio,
        "esperanza": df_tiradas["valor"].mean(),
        "varianza": df_tiradas["valor"].var(),
        "desvio_estandar": df_tiradas["valor"].std(),
        "df": df_tiradas,
    }
    
    print(f"Número de estudio: {numero_estudio}")
    print(f"Esperanza : {datos_tirada['esperanza']}")
    print(f"Varianza : {datos_tirada['varianza']}")
    print(f"Desviacion estandar : {datos_tirada['desvio_estandar']}")
    print("-------------------------------------------------")
    
    dict_tiradas[corrida+1] = datos_tirada
```

# CALCULO DE PROMEDIOS

```python
df_promedios = pd.DataFrame({"valor":dict_tiradas[1]["df"]["valor"], 
                             "frecuencia_relativa": dict_tiradas[1]["df"]["frecuencia_relativa"],
                             "esperanza": dict_tiradas[1]["df"]["esperanza"],
                             "varianza": dict_tiradas[1]["df"]["varianza"],
                             "desvio_estandar": dict_tiradas[1]["df"]["desvio_estandar"]
                            })
df_frec_promedios =  pd.DataFrame({"fa": frecuencias_por_corrida[0]["fa"], 
                             "fr": frecuencias_por_corrida[0]["fr"], 
                             "fr%": frecuencias_por_corrida[0]["fr"]
                            })
for corrida in range(1, cantidad_corridas):
    df_promedios["frecuencia_relativa"] += dict_tiradas[corrida+1]["df"]["frecuencia_relativa"]
    df_promedios["esperanza"] += dict_tiradas[corrida+1]["df"]["esperanza"]
    df_promedios["varianza"] += dict_tiradas[corrida+1]["df"]["varianza"]
    df_promedios["desvio_estandar"] += dict_tiradas[corrida+1]["df"]["desvio_estandar"]  
    
    df_frec_promedios["fa"] += frecuencias_por_corrida[corrida]["fa"]

df_promedios = df_promedios/cantidad_corridas
dict_promedios = {}
dict_promedios[1] = {
    "df": df_promedios
}

df_frec_promedios["fr"] = df_frec_promedios["fa"]/(cantidad_tiradas*cantidad_corridas)
df_frec_promedios["fr%"] = 100*df_frec_promedios["fr"]
```
